```python
def scrape_page(country_code, date):
  url = "https://spotifycharts.com/regional/" + country_code + "/weekly/" + date
  headers = {'user-agent': 'Chrome/76.67.59.152'}
  scraped = requests.get(url, headers = headers)
  scraped_html = bs(scraped.content, 'html.parser')

  df = pd.DataFrame(columns = ['Source', 'Position', 'Trend', 'Track', 'Streams'])
  
  table = scraped_html.find('table')

  if table:
    table_rows = table.find_all('tr')
    for row in table_rows:
      td = row.find_all('td')
      row_text = [str(t) for t in td]
      if row_text:
        new_row = pd.DataFrame({'Source': [row_text[0]],
                                'Position': [row_text[1]],
                                'Trend': [row_text[2]],
                                'Track': [row_text[3]],
                                'Streams': [row_text[4]]})
        df = df.append(new_row, ignore_index = True)
    
    df = df.assign(Country = countries[country_code], Country_Code = country_code, Date = date)

  return(df)
```

```python
spotify_data = pd.DataFrame()
for country_code in countries.keys():
  country_df = scrape_page(country_code, "2022-02-25--2022-03-04")
  spotify_data = spotify_data.append(country_df, ignore_index = True)
  print("Country Name: " + countries[country_code])
  print("Country Code: " + country_code)
  print("Scrapped: " + str(country_df.shape[0]) + " rows")
  time.sleep(1)
```
